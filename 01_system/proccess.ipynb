```bash
&         # Запуск в фоновом режиме => sleep 1000 &
fg        # Возобновление процесса
bg        # Возобновление процесса в фоне
jobs      # Список процессов в фоновом режиме
nohup     # Запрещает закрытие процесса в случае разрыва сессии
ps        # Текущее состояние процессов
   -a     # Все процессы
   -u     # Все пользователи
   -x     # Включая системные и т.д.
top       # Диспетчер процессов
free      # Количество оперативки и раздела подкачки
uptime    # Время работы с момента последнего включения
pgrep     # найти id's процессов по имени (default)
      -l  # Показывать имена
      -u  # Фильтр по пользователю
pkill     # Убить все процессы из списка id ^^^
kill      # Убить процесс
killall   # бить все процессы по имени
screen    # Утилита, сохраняющая сессии
Ctrl + Z  # Остановить процесс
```
