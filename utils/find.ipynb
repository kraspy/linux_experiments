## Основные команды
### Поиск файла по имени:
```bash
find /путь/до/директории -name "имя_файла"
```
### Поиск файлов с расширением `.txt`:
```bash
find /путь/до/директории -type f -name "*.txt"
```
### Поиск директорий:
```bash
find /путь/до/директории -type d
```
### Поиск файлов, измененных в течение последних 7 дней:
```bash
find /путь/до/директории -mtime -7
```
### Поиск файлов больше 100МБ:
```bash
find /путь/до/директории -size +100M
```
### Исключение сообщений об ошибках доступа (например, Permission denied):
```bash
find /путь/до/директории -type f -name "*.tmp" 2>/dev/null
```
### Вывод содержимого найденных файлов:
```bash
find /путь/до/директории -type f -exec cat {} + 2>/dev/null
  ```
  Это позволяет прочитать содержимое всех файлов, найденных командой `find`, игнорируя при этом сообщения об ошибках доступа.
## Примеры
### Поиск всех JPEG файлов в домашней директории и вывод их содержимого:
  ```bash
  find ~/ -type f -name "*.jpg" -exec cat {} + 2>/dev/null
  ```
### Удаление всех `.bak` файлов в текущей директории:
  ```bash
  find . -type f -name "*.bak" -exec rm {} +
  ```
