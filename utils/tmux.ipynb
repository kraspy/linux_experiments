```sh
Ctrl + B  # префиксная клавиша для доступа к командам tmux.
```

## Управление окнами
```sh
Ctrl + B c # создать новое окно.
Ctrl + B w # выбрать окно из списка.
Ctrl + B 0–9 # перейти к окну по его номеру.
Ctrl + B , # переименовать текущее окно.
Ctrl + B & # закрыть текущее окно.
```

## Управление панелями
```sh
Ctrl + B \"              # разделить текущее окно горизонтально.
Ctrl + B %               # разделить текущее окно вертикально.
Ctrl + B x               # закрыть текущую панель.
Ctrl + B стрелка         # перейти к соседней панели в указанном направлении.
Ctrl + B o               # перейти к следующей панели.
Ctrl + B ;               # переключиться между текущей и предыдущей панелью.
Ctrl + B Ctrl + стрелка  # изменить размер текущей панели.
```

## Управление сессиями
### Создание и подключение
```sh
tmux new-session -s <имя>                         # создать новую сессию с именем <имя>.
tmux attach-session -t <имя> или tmux a -t <имя>  # подключиться к существующей сессии.
tmux a                                            # вернуться в последнюю сессию после временного выхода.
```
### Отключение и завершение
```sh
Ctrl + B d                    # отключиться от текущей сессии (detach).
tmux detach-session -t <имя>  # отключить всех клиентов от сессии <имя>.
tmux kill-session -t <имя>    # завершить сессию <имя>.
```
tmux kill-server # остановить все запущенные сессии.
### Управление и информация
```sh
tmux list-sessions или tmux ls                   # вывести список всех запущенных сессий.
tmux has-session -t <имя>                        # проверить, существует ли сессия <имя>.
tmux rename-session -t <старое_имя> <новое_имя>  # переименовать сессию.
```

## Навигация и история
```sh
Ctrl + B [        # войти в режим копирования (просмотр истории).
Ctrl + B Page Up  # прокрутка вверх по истории.
Esc или Ctrl + C  # выйти из режима просмотра истории.
```

## Копирование и вставка
```sh
Ctrl + B ]  # вставить из внутреннего буфера обмена tmux.
```

## Командная строка tmux
```sh
Ctrl + B :  # открыть командную строку tmux для ввода команд напрямую.
```

## Полезные команды
```sh
tmux list-clients -t <имя_сессии>          # показать список клиентов, подключенных к сессии.
tmux new-session -s <имя> -c <директория>  # создать новую сессию с указанной стартовой директорией.
tmux detach                                # отключиться от текущей сессии.
tmux attach                                # подключиться к последней сессии.
```
