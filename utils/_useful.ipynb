## Text utils
Команды не редактируют исходный файл
```bash
cat         # Конкатенация файлов
cut         # Вырезает данные из файла (не редактирует исходный)
    -c 1,2  # Вывести символы по порядковым номерам
expand      # Преобразовать символы табуляции в пробелы
unexpand    # Преобразовать пробелы в табы
fmt         # Форматирование вывода текста
    -w 10   # Разместить 10 знаков текста в строке
head        # Показывает начало файла (10 строк по дефолту)
    -n 2    # Первые 2 строки
tail        # Показывает конец файла (10 строк по дефолту)
    -n 2    # Последние 2 строки
    -f      # Мониторить изменения в файле
join        # Объединение файлов по одинаковым ключам
paste       # Объединение файлов построчно
less        # Постраничное чтение файла
nl          # Нумерует строки
od          # Преобразование (восьмеричная система по дефолту)
     -c     # Преобразование к ASCII
pr          # Предпросмотр печати
sed         # Редактор текста (stream editor)
sort        # Сортирует строки по какому-либо признаку
     -r     # В обратном порядке
split       # Разбить файл
     -l 2   # По две строки
     -b 5   # По 5 байт
tr          # Трансформация
     -t     # Замена символов      -t A-Z a-z => hello
     -d     # Удаление символов    -d l       => Heo
     -s     # Объединение символов -s l       => Helo
wc          # Количество слов/строк/символ в файле
     *      # Для всех файлов в каталоге
     -w     # Количество слов
```
