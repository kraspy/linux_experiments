Обычно используется в сочетании с командой `sort`, так как `uniq` удаляет или отмечает только подряд идущие дубликаты строк.

## Основное использование
- Удаление дубликатов из отсортированного файла:
```bash
sort file.txt | uniq
```
- Подсчет количества каждой уникальной строки:
```bash
sort file.txt | uniq -c
```
- Вывод только тех строк, которые повторяются:
```bash
sort file.txt | uniq -d
```
- Вывод уникальных строк, исключая повторяющиеся:
```bash
sort file.txt | uniq -u
```
## Дополнительные опции
- `-c`: Подсчет количества вхождений каждой строки.
- `-d`: Вывод одного экземпляра каждой повторяющейся строки.
- `-u`: Вывод только тех строк, которые не повторяются.
- `-i`: Игнорирование различий в регистре символов при сравнении.
- `-s N`: Игнорирование первых N символов каждой строки при сравнении.
- `-w N`: Сравнение только первых N символов каждой строки.
- `-z`: Обработка ввода с нулевым символом в качестве разделителя строки (полезно при работе с именами файлов, содержащими новые строки).

## Примеры
### Удаление дубликатов с подсчетом их количества:
```bash
sort file.txt | uniq -c
```
Выведет количество и строку для каждой уникальной строки в файле.
### Нахождение повторяющихся строк без учета регистра:
```bash
sort file.txt | uniq -di
```
Покажет один экземпляр каждой повторяющейся строки, игнорируя регистр.
### Игнорирование первых N символов при сравнении строк:
```bash
sort file.txt | uniq -s N
```
Полезно, например, когда первые N символов каждой строки содержат временную метку или другие неуникальные данные.
