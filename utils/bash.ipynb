## Keymap
```bash
Ctrl + A         # Перейти в начало строки
Ctrl + E         # Перейти в конец строки
Ctrl + U         # Вырезать от начала строки до курсора 
Ctrl + K         # Вырезать от курсора до конца строки
Ctrl + W         # Вырезать слово слева (до разделителя)  
Alt + B          # Перейти на одно слово влево  
Alt + F          # Перейти на одно слово вправо  
Ctrl + R         # Инкрементальный поиск по истории назад  
Ctrl + D         # Если строка пустая — отправить EOF; иначе удалить символ под курсором  
Ctrl + L         # Очистить экран 
Ctrl + C         # Прервать текущий процесс (SIGINT)  
Ctrl + Z         # Приостановить текущий процесс (SIGTSTP), затем управлять через fg/bg  
Ctrl + Z => fg   # Возобновить приостановленную задачу на переднем плане  
Ctrl + Z => bg   # Возобновить приостановленную задачу в фоне  
app &            # Запустить команду app в фоне  
jobs             # Показать список задач текущего шелла (фон/приостановленные)  
fg %num          # Перевести задачу номер num на передний план  
bg %num          # Перевести задачу номер num в фон  
```

## Пути, Ввод/Вывод
```bash
.          # Текущая директория  
..         # Родительская директория (на уровень выше)  
~          # Домашняя директория текущего пользователя (tilde-расширение)

*          # любое количество символов (кроме /)  
?          # ровно один символ (кроме /)  
[a-z]      # любой один символ из диапазона/класса  
{1..5}     # Brace expansion: последовательность 1 2 3 4 5 (генерация слов до подстановок)  
f{1..2}    # Brace expansion: перечисление слов f1 f2 (генерация, не глоббинг)

< file     # Перенаправить stdin из файла  
> file     # Перенаправить stdout в файл (перезапись)  
>> file    # Перенаправить stdout в файл (дозапись)  
2> file    # Перенаправить stderr в файл (перезапись)  
2>> file   # Перенаправить stderr в файл (дозапись)  
|          # Конвейер: stdout слева → stdin справа
```

## env, 
```bash
env             # Показать окружение
A=123           # Создать (или изменить) переменную на время текущей сессии
echo "$A"       # Вывести значение переменной
export A        # Экспортировать переменную в окружение
unset A         # Удалить переменную

PATH="$PATH:."  # ДОБАВИТЬ текущую директорию в PATH

ln -s "$(pwd)/script" /usr/local/bin/script  # Создать симлинк

^mr^rm          # заменить первое вхождение 'mr' на 'rm' в ПРЕДЫДУЩЕЙ команде и выполнить.

alias gs='git status'   # Объявление алиаса
```

## Процессы
```bash
top            # Перечень запущенных процессов
top -u user    # Процессы пользователя user
ps             # Запущенные процессы
pgrep <pname>  # Поиск процесса по имени
kill <PID>     # Убивает процесс по PID
kill -9 <PID>  # Жестко убивает процесс по PID
xkill          # Клик по окну для завершения
killall <name> # Убивает все процессы по имени
time           # Время выполнения процесса
```
