## Структура файлов
```bash
/      # Корневой размер
/boot  # Загрузочный раздел
/bin   # Основные исполняемые файлы и утилиты
/sbin  # Основные исполняемые файлы и утилиты
/home  # Домашние директории пользователей
/dev   # Каталог файлов устройств
/root  # Домашняя директория суперпользователя
/etc   # Конфигурация системы и ее компонентов
/opt   # Директория ПО третьих поставщиков
/var   # Часто меняющиеся данные
/usr   # Пакеты программ, код ядра, документация
/lib   # Каталог с библиотеками и модулями ядра
/tmp   # Временные файлы
/proc  # Точка монтирования файловой системы procfs
/sys   # Точка монтирования файловой системы sysfs
swap   # не монтируется
```
## Прочее
```bash
tree             # Дерево файлов/каталогов
tree -d          # Дерево файлов/каталогов
ln file link     # Жесткая ссылка
ln -s file link  # Символическая ссылка
```
### Типы файлов
```bash
-  #
d  #
b  # блочное устройство
c  # символьное устройство
l  # символическая ссылка
s  # socket
p  # Именованный канал (pipe)
# chmod
chmod [-R] ugo+rwx file
```
### SUID
Запуск файла на исполнение с правами владельца
```bash
chmod u+s file
chmod 4755 file
```
### SGID
Работа в каталоге с правами группы владельца\
Применяется преимущественно к каталогам
```bash
chmod g+s catalog
chmod 2755 catalog
```
### Sticky bit
Файлы из каталога может удалить только их владелец\
Устанавивается только для каталогов
```bash
chmod +t catalog
chmod 1666 catalog
```
## Раздел подкачки
```bash
swapon      # Добавить/включить раздел подкачки
swapoff     # Отключить раздел подкачки
/etc/fstab  # Конфиг монтирования файловых систем
mkswap      # Создание раздела или файла swap
```
## fdisk
Утилита разбиения жесткого диска
```bash
fdisk -l        # Список дисков
fdisk /dev/sdb  # Работа с диском
```
## mkfs
Утилита создания файловой системы
```bash
mkfs.ext4 /dev/sdb1
```
